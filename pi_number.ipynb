```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

n_points = 10000
x = 2 * np.random.rand(n_points).astype(np.float64) - 1
y = 2 * np.random.rand(n_points).astype(np.float64) - 1

# Create figures
def create_figures():
    # Create circle
    circle=plt.Circle((0, 0), 1, color='blue', fill=False, linewidth=2)
    ax.add_artist(circle)
    ax.set_aspect('equal', adjustable='box')
    ax.set_xlim(-1.2, 1.2)
    ax.set_ylim(-1.2, 1.2)
    #Create square
    square = plt.Rectangle((-1, -1), 2, 2, color='red', fill=False, linewidth=1)
    ax.add_artist(square)
    plt.title('Generating pi number using Monte Carlo method')
    #Plot points
    ax.scatter(x, y, color='green', s=1)
    plt.show()

create_figures()

#Calculate pi number
def calculate_pi():
    inside_circle = (x**2 + y**2 <= 1)
    points_inside = np.sum(inside_circle)
    pi=4*points_inside/n_points
    return pi

calculate_pi()
```
